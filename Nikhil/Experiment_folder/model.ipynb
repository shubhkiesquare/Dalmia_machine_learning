```python
def minmax_scaler(df):
    
    scaled_df = pd.DataFrame()

    # Iterate over each column in the DataFrame
    for column in df.columns:
        # Calculate the minimum and maximum values of the column
        min_val = df[column].min()
        max_val = df[column].max()

        # Apply the MinMax scaling formula to scale the values
        scaled_values = (df[column] - min_val) / (max_val - min_val)

        # Add the scaled values to the scaled DataFrame
        scaled_df[column] = scaled_values
        #scaled_df = np.array(scaled_df)
    #scaled_df = scaled_df.values

    return scaled_df
```
